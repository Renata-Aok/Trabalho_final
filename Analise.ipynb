```python
# Convertendo as entradas (comentários) em uma sacola de palavras (bag of words). 
# Utilizando a vetorização de frequência relativa (CountVectorizer). 

from sklearn.feature_extraction.text import CountVectorizer

vetorizar = CountVectorizer(lowercase = False, max_features = 50)
bag_of_words = vetorizar.fit_transform(dados["comentario"])
```

```python
# O resultado da vetorização é uma matriz esparsa (grande quantidade de zeros espalhados pela matriz),
# onde cada coluna indica uma palavra e cada linha, uma entrada de texto, assim teremos uma matriz 
# esparsa 

bag_of_words
```

```python
# Visualizando a matriz esparsa - coluna indicando uma palavra e cada linha, uma entrada de texto

vetorizado = pd.SparseDataFrame(
    bag_of_words,
    columns = vetorizar.get_feature_names()
)
vetorizado
```

```python
# confirmando o formato dos dados após vetorização

print(bag_of_words.shape)
```

```python
# Dividir a base novamente com a base vetorizada

from sklearn.model_selection import train_test_split

treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words,
                                                              dados["categoria_num"],
                                                              random_state = 42)
```

```python
print(treino.shape)
print(teste.shape)
```

```python
from sklearn.linear_model import LogisticRegression

regressao_logistica = LogisticRegression(solver = "lbfgs", multi_class='auto')
regressao_logistica.fit(treino, classe_treino)
previsao_teste = regressao_logistica.predict_proba(teste)
print(previsao_teste)
```

```python
previsao_teste_bol = previsao_teste[:,1] >= 0.5
print(previsao_teste_bol)
```

```python
from sklearn.metrics import accuracy_score

acuracia_teste = accuracy_score(classe_teste, previsao_teste_int)
print(acuracia_teste)
```

```python
def nuvem_palavras_negativas(texto, coluna_texto):
    texto_negativo = texto.query("polaridade == 'negativo'")
    todas_palavras = ' '.join([texto for texto in texto_negativo[coluna_texto]])
    nuvem_palavras = WordCloud(width = 800, height = 500,
                                max_font_size=110, collocations=False).generate(todas_palavras)
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    
nuvem_palavras_negativas(dados, "comentario_1")
```

```python
def nuvem_palavras_positivas(texto, coluna_texto):
    texto_positivo = texto.query("polaridade == 'positivo'")
    todas_palavras = ' '.join([texto for texto in texto_positivo[coluna_texto]])
    nuvem_palavras = WordCloud(width = 800, height = 500,
                                max_font_size=110, collocations=False).generate(todas_palavras)
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    
nuvem_palavras_positivas(dados, "comentario_1")
```

```python
def pareto(texto, coluna_texto, quantidade):
    todas_palavras = ' '.join([texto for texto in texto[coluna_texto]])
    frequencia = nltk.FreqDist(token_espaco.tokenize(todas_palavras))
    df_frequencia = pd.DataFrame({"Palavra": list(frequencia.keys()),
                                 "Frequência": list(frequencia.values())})
    df_frequencia = df_frequencia.nlargest(columns = "Frequência", n = quantidade)
    plt.figure(figsize=(12,8))
    ax = sns.barplot(data = df_frequencia, x= "Palavra", y = "Frequência", color = 'gray')
    ax.set(ylabel = "Contagem")
    plt.show()   

pareto(dados, "comentario_1", 10)
```

```python
def classificar_texto(texto, coluna_texto, coluna_classificacao):
    vetorizar = CountVectorizer(lowercase=False, max_features=50)
    bag_of_words = vetorizar.fit_transform(texto[coluna_texto])
    treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words,
                                                                 texto[coluna_classificacao],
                                                                 random_state = 42)
    
    regressao_logistica = LogisticRegression(solver="lbfgs", multi_class='auto')
    regressao_logistica.fit(treino, classe_treino)
    
    return regressao_logistica.score(teste, classe_teste)

acuracia_teste = classificar_texto(dados, "comentario_1", "categoria_num")
print(acuracia_teste)
```
